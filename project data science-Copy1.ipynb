```python
import pandas as pd
from pgmpy.models import BayesianModel
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv('olympic_games.csv')

# Encoding categorical columns and discretizing numerical columns
label_encoder = LabelEncoder()
df['games_type_encoded'] = label_encoder.fit_transform(df['games_type'])
df['host_country_encoded'] = label_encoder.fit_transform(df['host_country'])
df['host_city_encoded'] = label_encoder.fit_transform(df['host_city'])
df['country_encoded'] = label_encoder.fit_transform(df['country'])

num_bins = 5 
df['athletes_binned'] = pd.cut(df['athletes'], bins=num_bins, labels=False)
df['teams_binned'] = pd.cut(df['teams'], bins=num_bins, labels=False)
df['year_binned'] = pd.cut(df['year'], bins=num_bins, labels=False)
df['competitions_binned'] = pd.cut(df['competitions'], bins=num_bins, labels=False)
df['gold_binned'] = pd.cut(df['gold'], bins=num_bins, labels=False)
df['silver_binned'] = pd.cut(df['silver'], bins=num_bins, labels=False)
df['bronze_binned'] = pd.cut(df['bronze'], bins=num_bins, labels=False)



model = BayesianModel([
    ('games_type_encoded', 'host_country_encoded'),
    ('host_city_encoded', 'host_country_encoded'),
    ('country_encoded', 'host_country_encoded'),
    ('athletes_binned', 'host_country_encoded'),
    ('teams_binned', 'host_country_encoded'),
    ('gold_binned', 'host_country_encoded'),
])

# Train model
model.fit(df, estimator=MaximumLikelihoodEstimator)

# inference
infer = VariableElimination(model)

# Example of querying the model
result = infer.query(variables=['host_country_encoded'], evidence={
    'games_type_encoded': 1,
    'host_city_encoded': 3,
    'country_encoded': 4,
    'athletes_binned': 2,
    'teams_binned': 1,
    'gold_binned':1
})

print(result)
```

```python
import pandas as pd
from scipy.linalg import svd

df = pd.read_csv('olympic_games.csv')

# Creating the Country-Medal Type Matrix
# Aggregating the total number of each type of medal for each country
medal_matrix = df.groupby('country')[['gold', 'silver', 'bronze']].sum()

# Performing SVD
U, s, VT = svd(medal_matrix)

# Results
print("U matrix:")
print(U[:5])  # Printing the first 5 rows of the U matrix

print("\nSingular values:")
print(s)  # Printing the singular values

print("\nVT matrix:")
print(VT[:5])  # Printing the first 5 rows of the VT matrix
```

```python
import numpy as np

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        # Initialize weights and biases
        self.W1 = np.random.randn(input_size, hidden_size)
        self.b1 = np.zeros(hidden_size)
        self.W2 = np.random.randn(hidden_size, output_size)
        self.b2 = np.zeros(output_size)

    def forward(self, X): # Forward pass
        self.z1 = np.dot(X, self.W1) + self.b1
        self.a1 = self.relu(self.z1)
        self.z2 = np.dot(self.a1, self.W2) + self.b2 
        self.a2 = self.softmax(self.z2)
        return self.a2

    def backward(self, X, y, learning_rate): # Backward pass
        m = X.shape[0]
        delta3 = self.a2 - y
        dW2 = np.dot(self.a1.T, delta3) / m
        db2 = np.sum(delta3, axis=0) / m
        delta2 = np.dot(delta3, self.W2.T) * self.relu_derivative(self.z1)
        dW1 = np.dot(X.T, delta2) / m
        db1 = np.sum(delta2, axis=0) / m
        # Update parameters
        self.W2 -= learning_rate * dW2
        self.b2 -= learning_rate * db2
        self.W1 -= learning_rate * dW1
        self.b1 -= learning_rate * db1

    def relu(self, x):
        # ReLU activation function
        return np.maximum(0, x)

    def relu_derivative(self, x):
        # Derivative of ReLU activation function
        return np.where(x > 0, 1, 0)

    def softmax(self, x):
        # Softmax activation function
        exp_scores = np.exp(x)
        return exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
```

```python
import numpy as np

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        # Initialize weights and biases
        self.weights_input_hidden = np.random.rand(input_size, hidden_size)
        self.bias_hidden = np.zeros((1, hidden_size))
        self.weights_hidden_output = np.random.rand(hidden_size, output_size)
        self.bias_output = np.zeros((1, output_size))
    
    def forward(self, X): # Forward pass
        self.hidden_activation = self.sigmoid(np.dot(X, self.weights_input_hidden) + self.bias_hidden)
        self.output = self.sigmoid(np.dot(self.hidden_activation, self.weights_hidden_output) + self.bias_output)
        return self.output
    
    def backward(self, X, y, learning_rate): # Backward pass
        error = y - self.output
        output_delta = error * self.sigmoid_derivative(self.output)
        hidden_error = output_delta.dot(self.weights_hidden_output.T)
        hidden_delta = hidden_error * self.sigmoid_derivative(self.hidden_activation)
        
        # Update weights and biases
        self.weights_hidden_output += self.hidden_activation.T.dot(output_delta) * learning_rate
        self.bias_output += np.sum(output_delta, axis=0, keepdims=True) * learning_rate
        self.weights_input_hidden += X.T.dot(hidden_delta) * learning_rate
        self.bias_hidden += np.sum(hidden_delta, axis=0, keepdims=True) * learning_rate
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def sigmoid_derivative(self, x):
        return x * (1 - x)

nn = NeuralNetwork(2, 3, 1)

# Training data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Training loop
epochs = 1000
learning_rate = 0.1

for epoch in range(epochs):
    # Forward pass
    output = nn.forward(X)
    
    # Backward pass
    nn.backward(X, y, learning_rate)
    
    # Print loss (mean squared error)
    loss = np.mean((output - y) ** 2)
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss = {loss:.4f}")

# Test the trained network
test_input = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
predicted_output = nn.forward(test_input)

print("Predicted Output:")
print(predicted_output)
```
